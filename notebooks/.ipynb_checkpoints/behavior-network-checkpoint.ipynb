```python
WII = 100
N = 500
s = 30
index = 35
for WII in np.arange(100,2000,100):
    for s in np.arange(0,100,10):
        index +=1
        data_name = "data-%d_WII-%d_N-%d_s-%d" %(index, WII, N, s)
        path = os.path.expanduser("/Users/GP1514/Dropbox/ICL-2014/Code/C-Code/cortex/scripts/%s.sh"%data_name)
        content = """#!/bin/bash
        N=8
        trap 'exit' INT
        num=$(awk 'BEGIN{for(i=0;i<4;i+=0.5)print i}')
        num2=$(awk 'BEGIN{for(i=0;i<100;i+=10)print i}')
        for n in $num
        do
          # ===============================================
          # Beginning of inner loop.
          (
          for j in $num2
          do
                ((i=i%%N)); ((i++==0)) && wait
                ../cortex/cortex -G $n -S $j -N %d -r 0.8 -d1 10 -d2 3000 -d3 10 -before 10 -after 3010 -s %d -WII %d -output _data_%d-wii%d-n%d-r08-s%d -model gp-izh &
          done
          )
          # End of inner loop.
          # ===============================================
        done
        # End of outer loop.
        exit 0
        """ % (N, s, WII,index, WII, N,s)

        with open(path,"w") as f:
            os.chmod(path, 0o770)
            f.write(content)

        print("start")
        subprocess.call([path])
        time.sleep(10)
        print("end")
```

```python
def plotAll(data_name):
    path = os.path.expanduser("/Users/GP1514/Dropbox/ICL-2014/Code/C-Code/cortex/data/%s"%data_name)
    column_names =['gammaC','nuEI','corI', 'spike','non-burst','burst', 'freq','power']
    df = pd.read_csv(path, names= column_names ,  dtype ='float32' , sep=";")
    df = df.fillna(10)
    # print(df.head())
    df= df[(df['nuEI']<200)]
    c=0
    figsize = (10,5)
    fig = plt.figure(figsize=figsize)
    gridsize=60
    ###################################
    # spiking
    ###################################
    dataframe = df[(df['freq']<1000)]
    title = r'Phase state diagram $spiking$'
    filename = "global_gap_junctions_c%d_spiking.pdf" %c
    column = 'non-burst'
    ax = fig.add_subplot(231)
    gr.plotDiagram(fig, ax, dataframe, title, column, filename, save=None, gridsize=gridsize, extent = [0,4,0,100] )

    ###################################
    # bursting
    ###################################
    dataframe = df[(df['freq']<1000)]
    title = r'Phase state diagram $bursting$'
    filename = "global_gap_junctions_c%d_bursting.pdf" %c
    dataframe['logburst'] =(dataframe['burst']).apply(np.log10)
    column = 'logburst'
    ax = fig.add_subplot(232)
    gr.plotDiagram(fig, ax, dataframe, title, column, filename, save=None, gridsize=gridsize, extent = [0,4,0,100] )

    ###################################
    # ratio
    ###################################
    df_sliced = df[(df['freq']<1000) & (df['burst']>0.00001)]
    df_sliced['ratio'] =((df_sliced['burst'])/df_sliced['non-burst']).apply(np.log10)
    dataframe = df_sliced
    title = r'Phase state diagram ratio $bursting/spiking$'
    filename = "global_gap_junctions_c%d_ratio.pdf"
    column = 'ratio'
    ax = fig.add_subplot(233)
    gr.plotDiagram(fig, ax, dataframe, title, column, filename, save=None, gridsize=gridsize, extent = [0,4,0,100] )

    # ###################################
    # # corI
    # ###################################
    # dataframe = df[(df['freq']<1000)]
    # title = r'Phase state diagram $corI$')
    # filename = "global_gap_junctions_c%d_corI.pdf" %c
    # column = 'corI'
    # fig = plt.figure(figsize=figsize)
    # ax = fig.add_subplot(111)
    # plotDiagram(fig, ax, dataframe, title, column, filename)

    ###################################
    # power
    ###################################
    dataframe = df[(df['freq']<1000)]
    ttl = r'Phase state diagram $P_dB(f)$'
    filename = "global_gap_junctions_c%d_power.pdf" %c
    dataframe['logpower'] =(dataframe['power']).apply(np.log10)
    column = 'logpower'
    ax = fig.add_subplot(234)
    gr.plotDiagram(fig, ax, dataframe, ttl, column, filename, save=None, front=False, gridsize=gridsize, extent = [0,4,0,100] )

    ###################################
    # power freq
    ###################################
    dataframe = df[(df['freq']<1000)]
    dataframe['logpower'] =(dataframe['power']).apply(np.log10)
    # dataframe = dataframe[(dataframe['logpower']>-1)]
    title = r'Phase state diagram $freq  Pdb>0.1$'
    filename = "global_gap_junctions_c%d_power_freq.pdf" %c
    column = 'freq'
    ax = fig.add_subplot(235)
    gr.plotDiagram(fig, ax, dataframe, title, column, filename, save=None,  front=False, gridsize=gridsize, extent = [0,4,0,100] )


    ax = fig.add_subplot(236)
    ax.set_title(data_name)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_axis_bgcolor((1, 1,1))
    ax.text(0.1,0.1, r'$W_{II} = %d$'%WII, fontsize=15, color='green')
    ax.text(0.1,0.5, r'$\sigma = %d$'%s, fontsize=15, color='red')
    plt.tight_layout()
    plt.savefig(DIRECTORY + 'all_%s.png'%data_name)
    plt.close(fig)
#         plt.show()
```
