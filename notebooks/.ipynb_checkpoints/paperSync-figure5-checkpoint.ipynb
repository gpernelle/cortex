```python
df = pd.DataFrame(columns=('nu','tauv','ratio','sG','N','k',
                           'argmax', 'max', 'm1', 'm2', 'mx',
                           'c1', 'c2', 'cx',
                           'f1', 'p1', 'f2', 'p2', 'end'
                          ) )
ratio = 0.5

i=0
N = 1000
dt = 0.1
T = 200000
s0 = 10000
d = 30000

nulist = [100]
tauvlist = np.arange(15,60,5)
sGlist = np.arange(0,100,10)
i=-1
mi = mutual_info
apple_ = generateInput2(3, T) * 20

sd0 = T//2-d
ed0 = sd0 + d

sc0 = T//2+2000
ec0 = sc0 + d
sc1 = T-d
ec1 = T

for k in [50, 100, 30, 10, 1]:
    for tauv in tauvlist:
        for sG in sGlist:
            for nu in nulist:
                for begin in [0, 1, 2]:
                    if begin==0:
                        s = sd0
                        e = ed0
                    elif begin==1:
                        s = sc0
                        e = ec0
                    else:
                        s = T-d
                        e = T
                    i+=1
    #                 rasterPath = "../data/rasters/rasterPlast8-tauv-%d_g-%d_N-%d_T-%d_k-%d_r-%.2f" % (tauv,sG, N, T, nu, ratio)
    #                 r = np.load(rasterPath)
                    varPath = "../data/rasters/rastervarPlastMI8-tauv-%d_g-%d_N-%d_T-%d_k-%d_r-%.2f_dt-%.2f_nu-%d" % (tauv, sG, N, T, k, ratio, dt , nu)
                    var = np.load(varPath)

                    i1 = var['i1N1'][s:e]
                    i2 = var['i1N2'][s:e]
                    apple = apple_[s:e]

                    f, Pxy = csd(i1, i2, nperseg=10000)

                    maxi = np.max(np.abs(Pxy))
                    argmax = np.argmax(np.abs(Pxy))
                    sigma = 0.01
                    norm = False
                    m1 = (mi.mutual_information_2d(apple, i1, 
                                                        normalized=norm, sigma=sigma))
                    m2 = (mi.mutual_information_2d(apple, i2, 
                                                        normalized=norm, sigma=sigma))
                    mx = (mi.mutual_information_2d(i1, i2,
                                                        normalized=norm, sigma=sigma))

                    c1 = np.corrcoef(apple, i1)[0,1]
                    c2 = np.corrcoef(apple, i2)[0,1]
                    cx = np.corrcoef(i1, i2)[0,1]
                    f1, p1 = fourier(i1)
                    f2, p2 = fourier(i2)
                    df.loc[i] = [nu, tauv, ratio, sG, N, k,
                                 argmax, maxi, 
                                 m1,m2,mx,
                                 c1,c2,cx,
                                 f1, p1, f2, p2, begin
                                 ]
```

```python
def facet_heatmap3(data, col='cor1', cols=['cor1', 'cor2', 'corChange'], **kws):
    #     data = pd.melt(df, id_vars=['tauv', 'sG'], value_vars=cols)
    data = data[data['variable'] == col]
    data = data.pivot(index='tauv', columns='sG', values='value')
    im = sns.heatmap(data, **kws)
    im.invert_yaxis()
    return im

def plotGrid(df, col, title='', cols=['cor1', 'cor2', 'corChange'], **kws):
    data = pd.melt(df, id_vars=['tauv', 'sG', 'end', 'net'], value_vars=cols)

    with sns.plotting_context(font_scale=5.5):
        g = sns.FacetGrid(data, col="end", row="net" )
    cbar_ax = g.fig.add_axes([.90, .3, .02, .4])
    g = g.map_dataframe(facet_heatmap3, col=col, cols=cols,cbar_ax=cbar_ax, **kws)
    for ax in g.axes.flat:
        ax.set_title("")
        ax.set_yticklabels([int(x) for x in tauvlist[::-1]])
        ax.set_xticklabels([int(x) for x in sGlist])
        
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle(title, fontsize='16')
    g.fig.subplots_adjust(right=.85)
#     g.savefig(PAPER + 'power-plot_%s.png' % col)

def plotGrid2(df, col, title='', cols=['cor1', 'cor2', 'corChange'], **kws):
    data = pd.melt(df, id_vars=['tauv', 'sG', 'end'], value_vars=cols)

    with sns.plotting_context(font_scale=5.5):
        g = sns.FacetGrid(data, col="end" )
    cbar_ax = g.fig.add_axes([.90, .3, .02, .4])
    g = g.map_dataframe(facet_heatmap3, col=col, cols=cols,cbar_ax=cbar_ax, **kws)
    for ax in g.axes.flat:
        ax.set_title("")
        ax.set_yticklabels(tauvlist[::-1])
        ax.set_xticklabels(sGlist)
        
    plt.subplots_adjust(top=0.9)
#     g.fig.suptitle(title, fontsize='16')
    g.fig.subplots_adjust(right=.85)
```

### gap junction strenght

```python
df = pd.DataFrame(columns=('tauv', 'sG', 'T', 'N', 'nu',
                          'csd_change',
                          'end', 'net',
                           'g','a',
                          )  )
i=0
N = 1000
dt = 0.1
T = 200000
s0 = 70000
d = 30000

nulist = [100]
tauvlist = np.arange(15,60,5)
sGlist = np.arange(0,100,10)

sc0 = T//2+2000
ec0 = T//2+12000
sc1 = T-10000
ec1 = T
k=0
ratio = 0.5
for nu in nulist:
    for tauv in tauvlist:
                for sG in sGlist:
                    for begin in [0,1,2]:
                        if begin==0:
                            s = sd0
                            e = ed0
                        elif begin==1:
                            s = sc0
                            e = ec0
                        else:
                            s = T-d
                            e = T
                        
                        for net in  ['N1', 'N2']:
                            i+=1
                            extension = extension = "-tauv-%d_g-%d_N-%d_T-%d_k-%d_r-%.2f_dt-%.2f_nu-%d" % (tauv, sG, N, T, k,  ratio, dt, nu)
                            filename = "../data/rasters/rastervarPlast8" + extension
                            if i==1: 
                                print(filename)
                            a = np.load(filename)
                            i1 = a['i1N1']
                            i2 = a['i1N2']
                            v1 = a['vvmN1']
                            v2 = a['vvmN2']
                            g1 = a['g1N1']
                            g2 = a['g1N2']
                            pN1 = a['pN1']
                            pN2 = a['pN2']
                            
                            v = v1 if net == 'N1' else v2
                            g = g1 if net == 'N1' else g2
                            p = pN1 if net == 'N1' else pN2
                            
                            mg = np.mean(g[s//100:e//100]) * (0.5*N-sG)/(0.5*N)
                            mp = np.mean(p[s:e]) 
                            mv = np.mean(v[s:e]) 
                            ma = mp/mv
                            
                            _, Pxy1 = csd(v1[sc1:ec1], v2[sc1:ec1], nperseg=1000)
                            _, Pxy0 = csd(v1[sc0:ec0], v2[sc0:ec0], nperseg=1000)
                            csd1 = np.max(np.abs(Pxy1)) 
                            csd0 = np.max(np.abs(Pxy0)) 
                            
                            csd_change = csd1 / csd0 
                                

                            df.loc[i] = [int(tauv), int(sG), int(T), int(N), int(nu),
                                        csd_change,
                                        begin, net!='N1',
                                        mg, ma]
```

```python
cols = ['g','a']
plotGrid(df[df['nu']==100], 'g', 'Gap-junction strength',cmap="magma", 
         cols = cols, )#vmin=0.005, vmax=0.01)
plt.savefig(PAPER + 'fig5-gamma.eps')

plotGrid(df[df['nu']==100], 'a', 'Activity',cmap="inferno", 
         cols = cols,)# vmin=0.48, vmax=0.60)
plt.savefig(PAPER + 'fig5-activity.eps')
```
