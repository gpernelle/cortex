```python
v= np.ones(30)*-45
u= np.zeros(30)
dt = 0.1
res_v = []
for t in np.linspace(0,1,100):
    v += dt * (1/20) * (v+55) * (v+40) - u + 100
    u += dt * 0.2 * (0.025*(v+55)**3)*(v>-55) - u
    v = -45*(v>25)+v*(v<=25)
    v = np.maximum(-60,v)
    res_v.append(np.mean(v))
```
