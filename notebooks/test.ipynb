```python
import numpy as np
import pylab as plt
%matplotlib inline
```

```python
plt.hist(np.random.normal(loc=30, scale=4, size=10000), bins=50)
```

```python
def normal(u,s,x):
    return 1/(s*(2*np.pi)**0.5)*np.exp(-(x-u)**2/(2*s**2))
```

```python
normal(30,4,30)
```

```python
t=0
N=1e5
for x in np.linspace(-100,100,N):
    t+=normal(30,4,x)/N
    
print(t)
```
